```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data1 = pd.read_csv('test.txt', sep=',', header=None)
plt.plot(data1[0], data1[1], label='x1')
plt.plot(data1[0], data1[2], label='x2')
plt.plot(data1[0], data1[3], label='x3')
plt.legend()
plt.xlabel('t')
plt.ylabel('Liczba czsteczek')
plt.grid()
plt.title('Pmax=1')
plt.savefig('Pmax=1.png')
plt.show()
```

```python
df = pd.read_csv('Pmax=5.txt', header=None, names=["run", "t", "x1", "x2", "x3"], sep=',')


color_map = {'x1': 'tab:blue', 'x2': 'tab:orange', 'x3': 'tab:green'}
shown_labels = set()

for run_id in sorted(df['run'].unique()):
    sub_df = df[df['run'] == run_id]

    for x_var in ['x1', 'x2', 'x3']:
        label = x_var if x_var not in shown_labels else None
        plt.plot(sub_df["t"], sub_df[x_var], color=color_map[x_var], alpha=0.6, label=label)
        shown_labels.add(x_var)

plt.xlabel("t")
plt.ylabel("Liczba czsteczek")
plt.title("Pmax=5")
plt.legend(loc='upper right')
plt.grid(True)
plt.tight_layout()
plt.savefig('Pmax=5.png')
plt.show()
```

```python
df_stat = pd.read_csv('Pmax=100_stat.txt', header=None, names=["t_bin", "x3_sr", "sigma_x3t"], sep=',')
plt.errorbar(df_stat["t_bin"], df_stat["x3_sr"], yerr=df_stat["sigma_x3t"], fmt='o-', markersize = 1)
plt.xlabel("t")
plt.ylabel("x3")
plt.title("Pmax=100 x3")
plt.grid(True)
plt.tight_layout()
plt.savefig('Pmax=100_stat.png')
plt.show()
```
