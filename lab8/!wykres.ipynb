```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
e3 = 'energy3.dat'
e4 = 'energy4.dat'
e5 = 'energy5.dat'
data_e3 = pd.read_csv(e3, header=None, delim_whitespace=True)
data_e4 = pd.read_csv(e4, header=None, delim_whitespace=True)
data_e5 = pd.read_csv(e5, header=None, delim_whitespace=True)

plt.plot(data_e3[0], data_e3[1], label='Podstawowe parametry')
plt.xlabel('Iteracja')
plt.ylabel('Energia [eV]')
plt.legend()
plt.grid()
plt.savefig('energy_plot3.png')
plt.show()
plt.plot(data_e4[0], data_e4[1], label='Korekta czterech wiza')
plt.xlabel('Iteracja')
plt.ylabel('Energia [eV]')
plt.legend()
plt.grid()
plt.savefig('energy_plot4.png')
plt.show()
plt.plot(data_e5[0], data_e5[1], label=r'Promie pocztkowy r$_\text{sr} = 2,5 \AA$')
plt.xlabel('Iteracja')
plt.ylabel('Energia [eV]')
plt.ylim([-450, 500])
plt.legend()
plt.grid()
plt.savefig('energy_plot5.png')
plt.show()
```

```python
r3 = pd.read_csv('radius3.dat', header=None, delim_whitespace=True)
r4 = pd.read_csv('radius4.dat', header=None, delim_whitespace=True)
r5 = pd.read_csv('radius5.dat', header=None, delim_whitespace=True)

plt.plot(r3[0], r3[1], label='Podstawowe parametry')
plt.xlabel('Iteracja')
plt.ylabel(r'r$_{sr}$ [$\AA$]')
plt.legend()
plt.grid()
plt.savefig('r_plot3.png')
plt.show()

plt.plot(r4[0], r4[1], label='Korekta czterech wiza')
plt.xlabel('Iteracja')
plt.ylabel(r'r$_{sr}$ [$\AA$]')
plt.legend()
plt.grid()
plt.savefig('r_plot4.png')
plt.show()

plt.plot(r5[0], r5[1], label=r'Promie pocztkowy r$_\text{sr} = 2,5 \AA$')
plt.xlabel('Iteracja')
plt.ylabel(r'r$_{sr}$ [$\AA$]')
plt.legend()
plt.grid()
plt.savefig('r_plot5.png')
plt.show()
```

```python
pcf3 = pd.read_csv('pcf_histogram3.dat', header=None, delim_whitespace=True)
plt.plot(pcf3[0], pcf3[1], label='Parametry podstawowe')
plt.xlabel('r [$\AA$]')
plt.ylabel('PCF [-]')
plt.legend()
plt.grid()
plt.savefig('pcf_plot3.png')
plt.show()

pcf4 = pd.read_csv('pcf_histogram4.dat', header=None, delim_whitespace=True)
plt.plot(pcf4[0], pcf4[1], label='Korekta czterech wiza')
plt.xlabel('r [$\AA$]')
plt.ylabel('PCF [-]')
plt.legend()
plt.grid()
plt.savefig('pcf_plot4.png')
plt.show()

pcf5 = pd.read_csv('pcf_histogram5.dat', header=None, delim_whitespace=True)
plt.plot(pcf5[0], pcf5[1], label=r'Promie pocztkowy $_\text{sr} = 2,5 \AA$')
plt.xlabel('r [$\AA$]')
plt.ylabel('PCF [-]')
plt.legend()
plt.grid()
plt.savefig('pcf_plot5.png')
plt.show()
```
