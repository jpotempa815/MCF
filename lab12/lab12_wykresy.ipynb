```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
import matplotlib.patches as patches
```

```python
os.chdir(r'C:\Users\julia\Desktop\studia\semestr 8\MCF\lab12\histogramy')
hists_zad1_r1 = ['hist_1r1100.dat', 'hist_1r1500.dat', 'hist_1r11000.dat', 'hist_1r110000.dat']
hists_zad1_r2 = ['hist_1r2100.dat', 'hist_1r2500.dat', 'hist_1r21000.dat', 'hist_1r210000.dat']

fig1, ax1 = plt.subplots(1, 4, figsize=([20,5]))
ax1 = ax1.flatten()
for i, hist in enumerate(hists_zad1_r1):
    data = pd.read_fwf(hist, header=None)
    ax1[i].plot(data[0], data[1], linewidth=3, label='Symulacja')
    ax1[i].plot(data[0], data[2], linewidth=3, linestyle='--', label='Teoria')
    ax1[i].set_title(hist.replace('.dat', ''), fontsize=26)
    ax1[i].set_yticks([])
    ax1[i].set_xlim(0, 1600)
    ax1[i].legend(fontsize=20)
    ax1[i].grid(True, alpha=0.2)
fig1.supxlabel(r'Prędkość [$\frac{m}{s}$]', fontsize=26)
fig1.supylabel('Liczba cząstek [-]', fontsize=26)
plt.subplots_adjust(left=0.4) 
plt.tight_layout()
plt.savefig('hists_zad1_r1.png', dpi=300)
plt.show()

fig2, ax2 = plt.subplots(1, 4, figsize=([20,5]))
ax2 = ax2.flatten()
for i, hist in enumerate(hists_zad1_r2):
    data = pd.read_fwf(hist, header=None)
    ax2[i].plot(data[0], data[1], linewidth=3, label='Symulacja')
    ax2[i].plot(data[0], data[2], linewidth = 3, linestyle='--', label='Teoria')
    ax2[i].set_title(hist.replace('.dat', ''), fontsize=26)
    ax2[i].set_yticks([])
    ax2[i].legend(fontsize=20)
    ax2[i].set_xlim(0, 1600)
    ax2[i].grid(True, alpha=0.2)
fig2.supxlabel(r'Prędkość [$\frac{m}{s}$]', fontsize=26)
fig2.supylabel('Liczba cząstek [-]', fontsize=26)
plt.subplots_adjust(left=0.4)  
plt.tight_layout()
plt.savefig('hists_zad1_r2.png', dpi=300)
plt.show()
```

```python
os.chdir(r'C:\Users\julia\Desktop\studia\semestr 8\MCF\lab12\wyniki')

nptv_3r1 = ['nptv_3r1100.dat', 'nptv_3r1500.dat', 'nptv_3r11000.dat', 'nptv_3r110000.dat']
nptv_3r2 = ['nptv_3r2100.dat', 'nptv_3r2500.dat', 'nptv_3r21000.dat', 'nptv_3r210000.dat']
nptv = nptv_3r1 + nptv_3r2
f = 16
nazwy = ['nptv_3r1100', 'nptv_3r1500', 'nptv_3r11000', 'nptv_3r110000', 'nptv_3r2100', 'nptv_3r2500', 'nptv_3r21000', 'nptv_3r210000']

for nptv_file in nptv:
    data = pd.read_fwf(nptv_file, header=None)
    fig, ax1 = plt.subplots(figsize=(10, 6))
    line1 = ax1.plot(data[0], data[3], color='tab:orange', marker='o', label='Temperatura')
    ax1.set_ylabel('Temeperatura (K)', fontsize=f)
    ax1.set_xlabel('Położenie w x [m]', fontsize=f)
    ax1.tick_params(axis='y')
    ax2 = ax1.twinx()
    ax2.set_ylabel('Ciśnienie (Pa)', fontsize=f)
    ax2.set_xlabel('Położenie w x [m]', fontsize=f)
    line2 = ax2.plot(data[0], data[2], color='tab:green', marker='s', label='Ciśnienie')
    ax2.tick_params(axis='y')
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right', fontsize=f)
    plt.tight_layout()
    plt.savefig(f'{nazwy[nptv.index(nptv_file)]}.png', dpi=300)
    plt.show()
```
