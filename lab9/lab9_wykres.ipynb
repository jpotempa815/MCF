```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

pliki_mc = ['pot_mc1.dat', 'pot_mc2.dat', 'pot_mc3.dat']
pliki_rel = ['pot_rel1.dat', 'pot_rel2.dat', 'pot_rel3.dat']
pliki_s = ['S_1.dat', 'S_2.dat', 'S_3.dat']
pliki_sigma = ['sigma_1.dat', 'sigma_2.dat', 'sigma_3.dat']



for mc, rel, s, sigma in zip(pliki_mc, pliki_rel, pliki_s, pliki_sigma):
    fig, axs = plt.subplots(1, 4, figsize=(20, 5))
    mc_data = np.loadtxt(mc)
    mc_data = mc_data.T
    s_data = np.loadtxt(s)
    s_data = s_data.T
    sigma_data = np.loadtxt(sigma)
    sigma_data = sigma_data.T
    rel_data = np.loadtxt(rel)
    rel_data = rel_data.T

    im0 = axs[0].imshow(mc_data, aspect='auto', cmap='inferno', origin='lower')
    im1 = axs[1].imshow(sigma_data, aspect='auto', cmap='cubehelix', origin='lower')
    im2 = axs[2].imshow(rel_data, aspect='auto', cmap='viridis', origin='lower')
    im3 = axs[3].imshow(s_data, aspect='auto', cmap='gnuplot', origin='lower')
    axs[0].set_xlabel('i')
    axs[0].set_ylabel('j')
    axs[0].set_title('a) Potencjał Monte Carlo')
    axs[1].set_title('b) Odchylenie standardowe')
    axs[2].set_title(r'c) |V_{rel} - V_{MC}|')
    axs[3].set_title('d) Zaabsorbowane łańcuchy')
    axs[1].set_xlabel('i')
    axs[2].set_xlabel('i')
    axs[3].set_xlabel('i')
    axs[0].grid()
    axs[1].grid()
    axs[2].grid()
    axs[3].grid()
    for ax, im in zip(axs, [im0, im1, im2, im3]):
        fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    fig.tight_layout()
    fig.savefig(f'{mc[:-4]}_wykresy.png')
    fig.show()

```
