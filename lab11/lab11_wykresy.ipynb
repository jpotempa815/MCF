```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
abs_names = ["abs_1.dat", "abs_2.dat", "abs_3.dat", "abs_4.dat", "abs_5.dat", "abs_6.dat", "abs_7.dat", "abs_8.dat"]
t_r_names = [("trans_1.dat", "refl_1.dat"), ("trans_2.dat", "refl_2.dat"), ("trans_3.dat", "refl_3.dat"), ("trans_4.dat", "refl_4.dat"), 
             ("trans_5.dat", "refl_5.dat"), ("trans_6.dat", "refl_6.dat"), ("trans_7.dat", "refl_7.dat"), ("trans_8.dat", "refl_8.dat")]
X = range(0, 101, 1)

for name_abs, tupl in zip(abs_names, t_r_names):
    df_abs = pd.read_csv(name_abs, header=None, delim_whitespace=True)
    plt.imshow(df_abs.T, aspect='auto', cmap='inferno', origin='lower', vmin = 0, vmax = 30)
    plt.colorbar()
    plt.xlabel('i [j.u.]')
    plt.ylabel('j [j.u.]')
    plt.savefig(name_abs.replace('.dat', '.png'))
    plt.show()
    for tup in tupl:
        plot_data = np.loadtxt(tup)
        plt.plot(X, plot_data, color='orange', linewidth=3)
        plt.xlim(0, 100)
        plt.xlabel('i [j.u.]')
        if 'refl' in tup:
            plt.ylabel('Reflektancja [j.u.]')
        else:
            plt.ylabel('Transmitancja [j.u.]')
        plt.grid()
        plt.savefig(tup.replace('.dat', '.png'))
        plt.show()
```
