```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
os.chdir(r"C:\Users\julia\Desktop\studia\semestr 8\MCF\lab13\wyniki")
files = ['nptv_200.dat', 'nptv_400.dat', 'nptv_600.dat', 'nptv_800.dat', 'nptv_1000.dat', 'nptv_1200.dat', 'nptv_1400.dat', 'nptv_1600.dat', 'nptv_1800.dat', 'nptv_2000.dat']
its = [200, 400, 600, 800, 1000, 1200, 1400, 1600, 1800, 2000]
rodzaje = [1, 2, 3, 4]
nazwy = ["gestosc", "cisnienie", "temperatura", "predkosc"]
nazwy2 = [r"Gęstość gazu $\left[\frac{\text{kg}}{\text{m}^3}\right]$", "Ciśnienie [Pa]", "Temperatura [K]", r"Prędkość gazu $\left[\frac{\text{m}}{\text{s}}\right]$"]
for rodzaj in  rodzaje:
        i = 0
        for file, it in zip(files, its):
            data=pd.read_fwf(file, header=None)
            plt.plot(data[0], data[rodzaj], label = f"it = {it}")
            plt.xlabel("Położenie x [m]", fontsize=14)
            plt.ylabel(f"{nazwy2[rodzaj-1]}", fontsize=14)
            plt.legend()
            plt.grid()
            if i == 4:
                plt.savefig(f"{nazwy[rodzaj-1]}1.png", dpi=300)
                plt.show()
            elif i == 9:
                plt.savefig(f"{nazwy[rodzaj-1]}2.png", dpi=300)
                plt.show()
            i+=1
```
